```python
import numpy as np
import tensorflow as tf
import re
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model
```

```python
# Load the IMDB dataset word index
word_index = imdb.get_word_index()
reverse_word_index = {value: key for key, value in word_index.items()}
```

```python
# Load the pre-trained model with ReLU activation
model = load_model('simple_rnn_imdb.keras')
model.summary()
```

```python
model.get_weights()
```

```python
# Function to preprocess user input
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"<br />", " ", text)
    text = re.sub(r"[^a-z\s]", "", text)  # remove punctuation

    words = text.split()
    encoded = [word_index.get(word, 2) for word in words]
    encoded = [i + 3 for i in encoded]

    padded = pad_sequences([encoded], maxlen=500)
    return padded
```

```python
# Prediction  function

def predict_sentiment(review):
    preprocessed_input=preprocess_text(review)
    prediction=model.predict(preprocessed_input)
    sentiment = 'Positive' if prediction[0][0] > 0.6 else 'Negative'
    return sentiment, prediction[0][0]
```

```python
# Example review for prediction

example_review = "the movie was bad, i don't like it much."
sentiment,score=predict_sentiment(example_review)
print(f'Review: {example_review}')
print(f'Sentiment: {sentiment}')
print(f'Prediction Score: {score}')
```
